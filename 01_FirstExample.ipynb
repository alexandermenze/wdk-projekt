## List Comprehension
... ist ein besonderes Feature von Python

Zwei einfache Beispiele:

```python
myLst = [1,2,3,4,6,9,12,22,33,56]
[lmt*3+7 for lmt in myLst]
```

List-Comprehension mit einer Funktion:

```python
def myFkt(param):
    return (param + 3 )/5

[myFkt(x) for x in myLst]
```

### Funktionen zur Notenberechnung
Konventionell:

```python
def note(punkte):
    for index, row in ndf.iterrows():
        if row['%'] <= punkte:  
            return row['Note']
    
note(50)    
```

pythonic:

```python
def noteB(punkte):
    return ndf[ndf['%'] <= punkte].iloc[0]['Note']
    
display(ndf[ndf['%'] < 67])
display(ndf[ndf['%'] < 67].iloc[0])
display(ndf[ndf['%'] < 67].iloc[0]['Note'])
```

```python
noteB(87)
```

### Anwendung der Notenberechnung auf die Spalte `pktProz`

```python
dfKerg['Note'] =  dfKerg['pktProz'].map(lambda proz: noteB(proz))
dfKerg.head()
```
